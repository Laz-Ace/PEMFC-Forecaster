### Function to take in model predictions and truth values and return evaluation

There are various metrics by wich we can evaluate the preformance of our model. Here we will make a function to evaluate our model with any relevant evaluation metric.

````python
def evaluate_preds(y_true, y_pred):
    # Calculate various evaluation metrics
    mae = tf.keras.metrics.mean_absolute_error(y_true, y_pred)
    mse = tf.keras.metrics.mean_squared_error(y_true, y_pred)
    rmse = tf.sqrt(mse)
    mape = tf.keras.metrics.mean_absolute_percentage_error(y_true, y_pred)
    
    # Account for different sized metrics
    if mae.ndim > 0:
        mae = tf.reduce_mean(mae)
        mse = tf.reduce_mean(mse)
        rmse = tf.reduce_mean(rmse)
        mape = tf.reduce_mean(mape)
    
    return {"mae": mae.numpy(), 
            "mse": mse.numpy(), 
            "rmse": rmse.numpy(), 
            "mape": mape.numpy()}
````

This portion of the notebook will be dedicated to windowing our dataset. This will turn our problem into a supervised learning problem.

[3, 4, 5, 6, 7, 8] -> [9]

````python
# Create function to label windowed data

def get_labelled_windows(x, horizon=1):
    """
    Creates labels for windowed dataset.
    """
    window = x[:, :-horizon]
    label = x[:, -horizon:]
    return window, label
````

Now that the windowing function is done, we need a method to iterate this function over the entire dataset.

The function used below is adapted from methods shown in this article:

https://towardsdatascience.com/fast-and-robust-sliding-window-vectorization-with-numpy-3ad950ed62f5

````python
import numpy as np
# function to view NumPy arrays as windows 
def make_windows(x, window_size=3, horizon=1):
    """
    Turns a 1D array into a 2D array of sequential windows of window_size.
    """
    # 1. Create a window of specific window_size (add the horizon on the end for later labelling)
    window_step = np.expand_dims(np.arange(window_size+horizon), axis=0)

    # 2. Create a 2D array of multiple window steps (minus 1 to account for 0 indexing)
    window_indexes = window_step + np.expand_dims(np.arange(len(x)-(window_size+horizon-1)), axis=0).T # create 2D array 
                                                                                                       # of windows of size 
                                                                                                       # window_size
    # 3. Index on the target array (time series) with 2D array of multiple window steps
    windowed_array = x[window_indexes]

    # 4. Get the labelled windows
    windows, labels = get_labelled_windows(windowed_array, horizon=horizon)

    return windows, labels
````

````python
# Make the train/test split but with our windowed data
def make_train_test_split(windows, labels, test_split=0.2):
    """
    Splits matching pairs of windows
    """
    split_size = int(len(windows) * (1 - test_split)) # Default 80% train 20% test
    train_windows = windows[:split_size]
    train_labels = labels[:split_size]
    test_windows = windows[split_size:]
    test_labels = labels[split_size:]
    return train_windows, test_windows, train_labels, test_labels
````

## Model Checkpoint

Machine learning invites a level of randomness that is not conducive to scientific comparison. For this reason, we will be comparing each models `"best"` preformance to all the others. This is to reduce variance in testing metrics. This next function will save the `"best"` model into a variable we can use later.

````python
import os

# Create a function to implement a ModelCheckpoint callback
def create_model_checkpoint(model_name, save_path="model_experiments"):
    return tf.keras.callbacks.ModelCheckpoint(filepath=os.path.join(save_path, model_name),
                                              monitor="val_loss",
                                              verbose=0,
                                              save_best_only=True)
````

## Making predictions into the future

````python
# How many timestamps to predict into the future.
INTO_FUTURE = 30
````

````python
 # 1. Create function to make predictions into future
def make_future_forecast(values, model, into_future, window_size=WINDOW_SIZE) -> list:
    """
    Make future predictions into_future steps after values end
    Return future forecasts as a list of floats
    
    values: Data being passed into model
    model: Model used for predicting
    into_future: Amount of timesteps into the future your model will predict
    window_size: The size of the window it will use
    """
        
    # 2. Make an empty list for future forecasts/prepare data to forecast on
    future_forecast = []
    last_window = values[-WINDOW_SIZE:] # only want preds from the last window (this will get updated)

    # 3. Make INTO_FUTURE number of predictions, altering the data which gets predicted on each time 
    for _ in range(into_future):
        # Predict on last window then append it again, again, again (model starts to make forecasts on its own forecasts)
        future_pred = model.predict(tf.expand_dims(last_window, axis=0))
        print(f"Predicting on: \n {last_window} -> Prediction: {tf.squeeze(future_pred).numpy()}\n")
    
        # Append predictions to future_forecast
        future_forecast.append(tf.squeeze(future_pred).numpy())
        # print(future_forecast)

        # Update last window with new pred and get WINDOW_SIZE most recent preds (model was trained on WINDOW_SIZE windows)
        last_window = np.append(last_window, future_pred)[-WINDOW_SIZE:]
  
    return future_forecast
````

````python
future_forecast = make_future_forecast(values=y_all,
                                       model=model_5,
                                       into_future=INTO_FUTURE,
                                       window_size=WINDOW_SIZE)
````

````python
def get_future_time (start_time, into_future, offset=1):
    """
    returns an array of seconds equal to start_time + into_future
    """
    start_time = start_time + (offset)
    end_time = start_time + into_future
    return np.arange(start_time, end_time)
````
